## 1. The Reflexive Market Loop

Unlike traditional ABMs where agents react to exogenous data, this simulation models **endogenous price discovery**.
- Agents receive the current price and recent news (Observables).
- They generate a `TradeDecision` (Buy/Sell/Hold with Volume).
- The `DataRegistry` aggregates net order flow.
- Market impact mechanics update the price algorithmically.
- The new price becomes the observation for the next step, creating a reflexive feedback loop.
